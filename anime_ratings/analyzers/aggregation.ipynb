```python
import os
import pandas as pd

pd.options.display.max_rows = 1000
home = os.path.expanduser('~')
path = f'{home}/Documents'

anime_ratings = pd.read_csv(f'{path}/s1vote.csv').set_index('title')
anime_meta = pd.read_csv(f'{path}/saraba1st.csv').set_index('title')
anime_fixed = pd.read_csv('./anime_fixed.csv').set_index('title')

df = anime_ratings.join(pd.concat([anime_meta, anime_fixed]))
df = df.reset_index().set_index('date')

df.sort_index().to_csv(f'{path}/anime-timeline.csv')
len(df)
```
